# working with json ( javacript object notation)

```python
json_string = '''
{
  "name": "Ram",
        "age": 25,
        "city": "Kathmandu",
        "email": "ram@example.com"
        }

'''
```

```python
import json
record = json.loads(json_string)
record 
```

```python
print(type(record))
```

```python
data =     {
        "name": "Sita",
        "age": 22,
        "city": "Lalitpur",
        "email": "sita@example.com"
}

  
```

```python
type(data)
```

```python
json_str = json.dumps(data)
json_str
```

```python
type(json_str)
```

```python
# creating new file 
with open ("output.json" , "w" ) as f : 
    json.dump(data , f , indent= 4)
```
