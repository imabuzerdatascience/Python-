# working with json ( javacript object notation)

```python
json_string = '''
{
  "name": "Ram",
        "age": 25,
        "city": "Kathmandu",
        "email": "ram@example.com"
        }

'''
```

```python
import json
record = json.loads(json_string)
record 
```

```python
print(type(record))
```

```python
data =     {
        "name": "Sita",
        "age": 22,
        "city": "Lalitpur",
        "email": "sita@example.com"
}

  
```

```python
type(data)
```

```python
json_str = json.dumps(data)
json_str
```

```python
type(json_str)
```

```python
# creating new file 
with open ("output.json" , "w" ) as f : 
    json.dump(data , f , indent= 4)
```

```python
with open ("product.json" , "r") as f :
   product = json.load( f)
product 
```

```python
type(product)
```

```python
Json_string = '''
{
  "books": [
    {
      "title": "The Magic Of Thinking Big",
      "author": "David J. Schwartz",
      "published": 1959
    },
    {
      "title": "Chanakya Neeti",
      "author": "B.K. Chaturvedi",
      "published": 2009
    },
    {
      "title": "How to Win Friends and Influence People",
      "author": "Dale Carnegie",
      "published": 1936
    }
  ],
  "status_code": 200
}
'''
```

```python
data = json.loads(Json_string)
data
```

```python
books = data["books"]
books
type(books)
```

```python
book_list = [{"title" : data["title"] , "author": data["author"]} for data in books ]
book_list
```
