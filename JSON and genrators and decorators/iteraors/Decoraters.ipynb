```python
#decorater 
def my_decorater(func) :
    def wrapper() :
        print("Function start hone wala hai ")
        func()
        print("function khatam hgaya hai ")
    return wrapper
```

```python
@my_decorater
def say_hello() :
    print("Hello  World")

say_hello()
```

```python
# decorater for sqyuare and cube number 
from functools import wraps

def square_cube_decorator(func):
    @wraps(func)
    def wrappers(*args, **kwargs):
        print("Calculating the square and cube")
        result = func(*args, **kwargs)
        for num, (square, cube) in result:
            print(f"Number: {num}, Square: {square}, Cube: {cube}")
        print("Calculation complete!")
        return result
    return wrappers

    
            
```

```python
@square_cube_decorator
def compute_square_cube(numbers):
    return [(num, (num**2, num**3)) for num in numbers]

# Example usage
compute_square_cube([2, 3, 4])

  
```
