```python
import pandas as pd 
df = pd.DataFrame({
    "Quarter": ["Q1 2022", "Q2 2022", "Q3 2022", "Q4 2022", "Q1 2023"],
    "Fridge": [12.5, 11.0, 13.2, 16.4, 13.8],
    "Dishwasher": [8.0, 7.5, 9.0, 11.0, 9.5],
    "Washing Machine": [9.2, 8.5, 7.5, 7.6, 6.8]
})
df
```

```python
from matplotlib import pyplot as plt 
plt.figure(figsize =(12,4)) # for wider size 
plt.plot(df["Quarter"] , df["Fridge"])
```

```python
plt.plot(df["Quarter"],df["Washing Machine"])
plt.show()
```

```python
plt.figure( figsize=(12,4))
plt.plot(df["Quarter"] , df['Dishwasher'] , color= "red" , label="Dishwasher")
plt.plot(df["Quarter"] , df['Washing Machine'] , color= "pink" , label="Washing machine")
plt.plot(df["Quarter"] , df['Fridge'] , color= "black" , label="Fridge")
plt.title("Product sales")
plt.ylabel('rev(mins)')
plt.xlabel('year')
plt.legend()
```

```python
Total_sales = df[["Fridge" ,"Washing Machine" , "Dishwasher"]].sum()
Total_sales
```

```python
Total_sales.index
```

```python
# lets plot a pie chart 
plt.pie(Total_sales , labels= Total_sales.index , autopct="%1.1f%%" , explode=(0.1, 0 , 0.1) , shadow=True , startangle=180 )
plt.show()
```

```python
# lets plot a barchart 
plt.bar(df["Quarter"], df["Dishwasher"] , color="green" , width=0.4)
plt.title("Product_list")
plt.xticks(rotation =45)
plt.yticks(rotation = 35)
plt.show()
```

```python
# histogram 
plt.hist(df["Washing Machine"])
```
