```python
# without list comprehensive 
numbers = [2,4,6,8,10]
square = []
for  n in numbers : 
    square.append(n*n)
print(square)
```

```python
# using list comprehensive 
square = [ n*n for n in numbers ]
square
```

```python
# another problem 
price = [100 , 200 , 180 , 150 ]
earing_per_share = [9 , 15 , 40 , 20 ]

# zip  
tuple(zip(price , earing_per_share))
```

```python
pe = [] 
for p , e in zip(price , earing_per_share) : 
    pe.append(p/e)
print(pe)
```

```python
# using list comprehensive 
pe = [p/e for p,e in zip(price , earing_per_share)]
pe
```

```python
# what if earing share is zero 
price = [20 , 30 , 40 ,50 ]
earing_share = [0 , 2 , 4, 6 ] 
# then we used if and else for 0 
pe = [p/e if e != 0 else -1 for p,e in zip(price , earing_share)]
pe
```
