```python
# starting with string
first = "Abuzer"
last = "khan"
print(first)
print(last)
```

```python
# we can accest the letter of the name using indexing number
first[0]
first[1]
first[2]
```

```python
# slicing method it helps to accest the letter form start to end or middle to ending
first[0:2] #it is slicing method
last[0 : 5] 
first[1:]
# negative slicing 
lastname = "python"
lastname[-6:]
lastname[-4 : -2]
lastname[ :: -1]
```

```python
# write full name using python 
# it is a concatination to add two string 
fullname = first + " " + last 
print(fullname)
```

```python
# another way to write full name ( python format string )
age = 23
fullname2 = f"{first} {last}  is {age} years old "
print(fullname2)
```

```python
#strip function it only remove space between from the beginning and end of a string. It does not remove spaces in the middle of the text.
Name = 'Mohammad Abuzer khan'
age = 23 
address = " Biratnagar-06".strip()
Degree = 'Bit'
role = " Data Scientist".strip()
Aboutmyself = f''' {Name} is {age} years old . He is living in {address} and he had passed bachelor in {Degree} and its work as a {role}. '''

print(Aboutmyself)
```

```python
#replace function 
Name = 'Mohammad Abuzer khan'
age = 23 
address = " Biratnagar-06"
Degree = 'Bit'
role = " Data Scientist"

Aboutmyself = f'''{Name} is {age} years old. He is living in {address} and he had passed bachelor in {Degree} and works as a {role}.'''

print(Aboutmyself.replace("  " , " " ))
```
