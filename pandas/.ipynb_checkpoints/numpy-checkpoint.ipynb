```python
# numpy is much faster than list 
import numpy as np 
import time 
size = 10000
l1 = range(size)
l2 = range(size)
start_time = time.time()
l3 = [ x+y for x,y in zip(l1,l2)]
end_time = time.time()
print("python list took : " , end_time - start_time ) 


L1 = np.arange(size)
L2 = np.arange(size)
start_time = time.time()
L3 = L1 + L2 
end_time = time.time()
print("Numpy list took : " , end_time - start_time ) 
```

```python
# matrix operation 
import numpy as np 
first_1 = np.array([[10,12,14],[15,16,18],[20,22,24]])
first_2 = np.array([[14,16,14],[17,19,21],[22,24,26]])
combine = first_1 + first_2
print(combine)
# find first_2 sale compare to q1 
compare = first_2 - first_1
print(compare)
# get percentage  of first_2 profit 
percent = ((first_2 - first_1) / first_1 ) * 100
print(percent)
# FIND REV 
prices = np.array([[10 ,20, 15],[6,8,19],[5,7,2]])
print(prices.shape)
Q1_rev = first_1 + prices
print(f" The Q1 rev is: {Q1_rev}" )
# find discount if owner gives 20 percent discount on actual price 
discount = Q1_rev * 0.2
print(f"The discount amount is {discount}")
# find net rev
Netrev =  Q1_rev - discount 
print(f" The net rev is : {Netrev}")
```

```python
# combine two data frame
coustomer_Deatils = np.array([
                    [101 , "Mira"] ,
                    [102, "saqib"] ,
                    [103, "bakar"]])
coustomer_More_Deatils = np.array([
                    [101 , 250.50 , '2023-08-01'] ,
                    [102, 350.50 , '2024-08-01'] ,
                    [103 ,150.50 , '2021-08-01'] ])
Coustomer = np.array([
                    [104 , "ira"] ,
                    [105, "aqib"] ,
                    [106, "akar"]])

Combine_horizontal = np.hstack((coustomer_Deatils , coustomer_More_Deatils))
Combine_verticallty = np.vstack((coustomer_Deatils , Coustomer))
print(Combine_horizontal)
print(Combine_verticallty)
```
