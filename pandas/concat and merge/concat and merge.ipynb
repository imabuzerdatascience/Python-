```python
Player_two =  pd.DataFrame({
     "scores" : [ 21, 29,34,56] ,
     "player" : ['mannu' , 'motu' , 'chotu' , 'majid']
    }) 
Player_two
```

```python
concat = pd.concat([Player_one , Player_two])
concat 
```

```python
df = pd.concat([Player_one , Player_two] , keys = [ "India" , "Nepal"] )
df
```

```python
# merge 
Player_one =  pd.DataFrame({
     "scores" : [ 22, 24,25,26] ,
     "player" : ['sachin' , 'rahul' , 'danish' , 'subash']
    }) 
Player_one
```

```python
Player_two =  pd.DataFrame({
     "scores" : [ 22, 29,34,56] ,
     "player" : ['sachin' , 'motu' , 'chotu' , 'majid']
    }) 
Player_two
```

```python
df = pd.merge( Player_one , Player_two , on ="scores") # ( on) find to common link between two value
df
```

```python
df = pd.merge( Player_one , Player_two , how = "left")
df
```

```python
df = pd.merge( Player_one , Player_two , how = "right")
df
```

```python
df = pd.merge( Player_one , Player_two , how = "outer")
df
```

```python
df = pd.merge( Player_one , Player_two , on="player" )
df
```
